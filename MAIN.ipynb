```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/MILLENIUM PRO-2/FIFA 2018 Statistics.csv')
print(df.head())
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Categorical Columns:")
print(categorical_columns)
print("\nNumerical Columns:")
print(numerical_columns)
for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
df[numerical_columns].describe()
```

```python
team_count = df['Team'].nunique()
print(f"Total number of unique teams: {team_count}")
```

```python
first_10_rows = df.head(10)
print("First 10 rows:")
print(first_10_rows)
last_10_rows = df.tail(10)
print("\nLast 10 rows:")
print(last_10_rows)
```

```python
man_of_match_counts = df.groupby('Team')['Man of the Match'].count()
print("Total count of 'Man of the Match' from each team:")
print(man_of_match_counts)
```

```python
opponents_list = df['Opponent'].unique()
print("List of unique opponents:")
for opponent in opponents_list:
    print(opponent)
```

```python
top_teams = df.sort_values('Pass Accuracy %', ascending=False)['Team'].head(10)
print("Top 10 teams with highest accuracy:")
for i, team in enumerate(top_teams, 1):
    print(f"{i}. {team}")
```

```python
lowest_teams = df.sort_values('Pass Accuracy %', ascending=True)[['Team', 'Ball Possession %']].head(5)
print("Teams with the lowest 5 accuracy and their ball possession:")
print(lowest_teams)
```

```python
average_ball_possession = df['Ball Possession %'].mean()
print(f"Average Ball Possession: {average_ball_possession}")
```

```python
average_ball_possession = df.groupby('Team')['Ball Possession %'].mean()
print("Average Ball Possession for each team:")
print(average_ball_possession)
```

```python
max_on_target = df['On-Target'].max()
min_on_target = df['On-Target'].min()
average_on_target = df['On-Target'].mean()

print(f"Max On-Target: {max_on_target}")
print(f"Min On-Target: {min_on_target}")
print(f"Average On-Target: {average_on_target}")
```

```python
max_off_target = df['Off-Target'].max()
min_off_target = df['Off-Target'].min()
average_off_target = df['Off-Target'].mean()

print(f"Max Off-Target: {max_off_target}")
print(f"Min Off-Target: {min_off_target}")
print(f"Average Off-Target: {average_off_target}")
```

```python
yellow_cards_analysis = df.groupby('Team')['Yellow Card'].sum()
print("Total Yellow Cards for each team:")
print(yellow_cards_analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
yellow_cards_analysis = df.groupby('Team')['Yellow Card'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=yellow_cards_analysis.index, y=yellow_cards_analysis.values)
plt.title('Total Yellow Cards for Each Team')
plt.xlabel('Team')
plt.ylabel('Yellow Card')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=yellow_cards_analysis.index, y=yellow_cards_analysis.values)
plt.title('Total Yellow Cards for Each Team')
plt.xlabel('Team')
plt.ylabel('Yellow Card')
plt.xticks(rotation=90)
plt.show()
```

```python
print(df.groupby('Team')['Passes'].describe())

# Grouped Analysis
print(df.groupby(['Team', 'Opponent'])['Passes'].mean())

# Correlation Analysis
print(df['Man of the Match'].dtype)
print(df.corr())
# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Passes', data=df)
plt.title('Average Passes by Team')
plt.show()
```

```python
average_free_kicks = df.groupby('Team')['Free Kicks'].mean()

# Print the average free kicks for each team
print("Average Free Kicks for each team:")
print(average_free_kicks)

# For a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=average_free_kicks.index, y=average_free_kicks.values)
plt.title('Average Free Kicks for Each Team')
plt.xlabel('Team')
plt.ylabel('Free Kicks')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility if needed
plt.show()

# For a line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_free_kicks.index, y=average_free_kicks.values)
plt.title('Average Free Kicks for Each Team')
plt.xlabel('Team')
plt.ylabel('Free Kicks')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility if needed
plt.show()
```

```python
correlations = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
sns.pairplot(df)
plt.show()
```

1.Line Chart: This is useful for visualizing a trend in data over intervals of time - a time series. 2.Bar Chart: This is useful when you want to compare a single category of data between individual sub-items. 3.Histogram: This is useful for visualizing the underlying frequency distribution (shape) of a set of continuous data. 4.Scatter Plot: This is useful for visualizing relationships between two numerical variables where one variable is independent. 5.Pie Chart: This is useful when you want to visualize a categorical dataset and see the proportion of each category in the data. 6.Box Plot: This is useful for visualizing the statistical summary of a data set including minimum, first quartile, median, third quartile, and maximum. 7.Heatmap: This is useful for visualizing correlation between multiple variables in a dataset. 8.Pair Plot: This is useful for visualizing pairwise relationships in a dataset. 9.Area Chart: This is useful for visualizing quantitative data graphically.
